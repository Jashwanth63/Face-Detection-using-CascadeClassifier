```python
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_xml = cv2.CascadeClassifier('/home/lawliet/anaconda3/lib/python3.7/site-packages/cv2/data/haarcascade_frontalface_alt.xml')
```

```python
Mface1 = cv2.imread('/home/lawliet/Documents/ML/ML/data/Faces/Male_face.jpg')
Mface1_gray = cv2.cvtColor(Mface1, cv2.COLOR_BGR2GRAY)
Fface1 = cv2.imread('/home/lawliet/Documents/ML/ML/data/Faces/Female_face.jpg')
Fface1_gray = cv2.cvtColor(Fface1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(Mface1)
#plt.imshow(Fface1, cmap='gray')
```

```python
plt.imshow(Mface1, cmap ='gray')
```

```python
Male_face = face_xml.detectMultiScale(Mface1_gray)
Female_face = face_xml.detectMultiScale(Fface1_gray)
```

```python
for (x1,y1,w1,h1) in (Male_face):
    cv2.rectangle(Mface1, (x1,y1), (x1+w1, y1+h1), (255,0,255), 10)
    Detected_male = Mface1[ y1:y1+h1+10, x1: x1+w1]
for (x2,y2,w2,h2) in (Female_face):
    cv2.rectangle(Fface1, (x2,y2), (x2+w2, y2+h2), (255,0,255), 10)
    Detected_female = Fface1[y2: y2+h2+10, x2: x2+w2]
    
cv2.imshow('Male',Mface1)
cv2.imwrite('Male.png', Detected_male)
cv2.imwrite('Female.png', Detected_female)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
